```python
df = pd.read_csv('HAM10000_metadata.csv')
```

```python
df['path'] = df['image_id'].map(imageid_path_dict.get)
df['cell_type'] = df['dx'].map(lesion_type_dict.get)
df['cell_type_idx'] = pd.Categorical(df['cell_type']).codes
```

```python
import cv2

# Function to read and resize images
def read_and_resize_image(image_path, size):
    try:
        img = cv2.imread(image_path)
        img = cv2.resize(img, size)
        return img
    except Exception as e:
        print(f"Error processing image: {image_path}")
        print(e)
        return None

image_size = (64, 64)
# Apply the function to each image path in the 'path' column of the dataframe
df['image'] = df['path'].apply(lambda x: read_and_resize_image(x, (64,64)))
```

```python
shape_counts = df['image'].dropna().map(lambda x: x.shape).value_counts()
missing_count = df['image'].isnull().sum()
shape_counts[(64, 64, 3)] += missing_count
print(shape_counts)
```

```python
# Split the data into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['cell_type_idx'])
```

```python
image_shapes = train_df['image'].dropna().map(lambda x: x.shape)
print(image_shapes)
```

```python
# Extract the image data and target labels for train and test sets
X_train = np.stack(train_df['image'].dropna().values)
y_train = train_df['cell_type_idx'].dropna().values
X_test = np.stack(test_df['image'].dropna().values)
y_test = test_df['cell_type_idx'].dropna().values
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, LSTM, Reshape
from tensorflow.keras.models import Model

# Define input shape

input_shape = (64, 64, 3)
# Input layer
inputs = Input(shape=input_shape)

# Convolutional layers
x = Conv2D(32, (3, 3), activation='relu')(inputs)
x = MaxPooling2D((2, 2))(x)

x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)

# Flatten layer
x = Flatten()(x)

# Reshape the tensor to add a time dimension
x = Reshape((1, -1))(x)
#By adding the Reshape layer with (1, -1) as the shape argument, 
#we effectively add a time dimension to the tensor.
#The -1 indicates that the size of the remaining dimensions is automatically determined based on the input shape.


# LSTM layer
x = LSTM(128)(x)
# Output layer
outputs = Dense(len(df['cell_type_idx'].unique()), activation='softmax')(x)

# Create the model
model = Model(inputs=inputs, outputs=outputs)

model.summary()
```

```python
model.compile(optimizer = 'adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=64, epochs=100, validation_split=0.2)
```

```python
# Create a new DataFrame with 'lesion_id' and 'cell_type_dx' columns from the test data
result_df = pd.DataFrame({'lesion_id': test_df['lesion_id'], 'target': y_test})

# Save the DataFrame to a CSV file
result_df.to_csv('submission.csv', index=False)
```
